```python
import matplotlib.pyplot as plt
```

```python
# T=15

# first plot: total maintenance cost of PSEM
x1 = [3, 4, 5, 6, 7, 8, 9]  
y1_1 = [200, 200, 220, 300, 310, 400, 400]   # cost of new strategy
y1_2 = [120, 130, 230, 330, 430, 440, 450]   # cost of naive strategy

# second plot: total leakage quantity of PSEM
x2 = [3, 4, 5, 6, 7, 8, 9]
y2_1 = [3.499, 4.967, 6.176, 7.499, 8.71, 8.766, 11.234] # leakage of new strategy
y2_2 = [6.0856, 8.1753, 10.6992, 13.2231, 16.0398, 18.1461, 20.236] # leakage of old strategy

# create figure and axis objects
fig, axs = plt.subplots(1, 2, figsize=(12, 6))  

# first plot
axs[1].plot(x1, y1_1, marker='o', linestyle='-', color='r', label='New Strategy')
axs[1].plot(x1, y1_2, marker='o', linestyle='-', color='b', label='Old Strategy')

axs[1].set_title("Total Maintenance Cost of PSEM (T=15)")
axs[1].set_xlabel("Number of CSEM")
axs[1].set_ylabel("Total Maintenance Cost")
axs[1].legend()

# second plot
axs[0].plot(x2, y2_1, marker='x', linestyle='-', color='r', label='New Strategy')  
axs[0].plot(x2, y2_2, marker='x', linestyle='--', color='b', label='Old Strategy')  

axs[0].set_title("Total leakage quantity of PSEM (T=15)")
axs[0].set_xlabel("Number of CSEM")
axs[0].set_ylabel("Leakage Quantity (kg)")
axs[0].legend()

plt.tight_layout()

plt.show()
```
