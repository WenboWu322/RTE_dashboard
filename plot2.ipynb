```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# T=15

# first plot: total maintenance cost of PSEM
x1 = [3, 4, 5, 6, 7, 8, 9]  
y1_1 = [200, 200, 220, 310,	320, 410, 420]   # cost of new strategy
y1_2 = [120, 130, 230, 330, 430, 440, 450]   # cost of naive strategy

# second plot: total leakage quantity of PSEM
x2 = [3, 4, 5, 6, 7, 8, 9]
y2_1 = [3.411, 4.987, 6.24, 6.692, 8.678, 9.998, 10.868] # leakage of new strategy
y2_2 = [6.0404, 8.1291, 10.6075, 13.0859, 15.8518, 17.9504, 20.0394] # leakage of old strategy

# create figure and axis objects
fig, axs = plt.subplots(1, 2, figsize=(12, 6))  

# first plot
axs[1].plot(x1, y1_1, marker='o', linestyle='-', color='r', label='New Strategy')
axs[1].plot(x1, y1_2, marker='o', linestyle='-', color='b', label='Old Strategy')

axs[1].set_title("Total Maintenance Cost of PSEM (T=15)")
axs[1].set_xlabel("Number of CSEM")
axs[1].set_ylabel("Total Maintenance Cost")
axs[1].legend()

# second plot
axs[0].plot(x2, y2_1, marker='x', linestyle='-', color='r', label='New Strategy')  
axs[0].plot(x2, y2_2, marker='x', linestyle='--', color='b', label='Old Strategy')  

axs[0].set_title("Total leakage quantity of PSEM (T=15)")
axs[0].set_xlabel("Number of CSEM")
axs[0].set_ylabel("Leakage Quantity (kg)")
axs[0].legend()

plt.tight_layout()

plt.show()
```

```python
# heatmap of total maintenance cost of PSEM (new strategy)
# 数据矩阵
data = np.array([
    [200, 200, 220, 310, 320, 410, 420],    # 对应15天
    [210, 220, 320, 330, 350, 440, 450],   # 对应30天
    [220, 240, 340, 360, 380, 480, 500]  # 对应45天
])

# 行和列标签
row_labels = ['15天', '30天', '45天']
col_labels = ['3', '4', '5', '6', '7', '8', '9']

# 绘制热力图
plt.figure(figsize=(4,3))
plt.imshow(data, cmap='YlGnBu', aspect='auto') # 使用渐变色

# 添加刻度和标签
plt.xticks(ticks=np.arange(len(col_labels)), labels=col_labels)
plt.yticks(ticks=np.arange(len(row_labels)), labels=row_labels)

# 添加颜色条
cbar = plt.colorbar()
cbar.set_label('Total Maintenance Cost')

# 在每个单元格中显示数值
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        plt.text(j, i, f'{data[i, j]}', ha='center', va='center', color='black')

plt.title('不同CSEM数量与时空长度下的泄漏量热力图')
plt.xlabel('CSEM number')
plt.ylabel('Horizon')
plt.tight_layout()
plt.show()
```
